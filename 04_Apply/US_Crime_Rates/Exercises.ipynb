### Step 3. Assign it to a variable called crime.

```python
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv"
crime = pd.read_csv(url)
crime.head()
```

### Step 4. What is the type of the columns?

```python
crime.dtypes
```

##### Have you noticed that the type of Year is int64. But pandas has a different type to work with Time Series. Let's see it now.

### Step 5. Convert the type of the column Year to datetime64

```python
crime['Year'] = pd.to_datetime(crime['Year'], format="%Y")
crime.head()
```

### Step 6. Set the Year column as the index of the dataframe

```python
crime.set_index('Year', inplace=True)
crime.head()
crime2 = crime.copy()
```

### Step 7. Delete the Total column

```python
crime.drop('Total', axis=1)
crime.head()
```

### Step 8. Group the year by decades and sum the values

#### Pay attention to the Population column number, summing this column is a mistake

```python
# Could have also used resample with 10AS option for every 10 years, start
decade = 10* (crime.index.year//10)
crime_decade = crime.drop("Population", axis=1).groupby(decade).sum()
#crime.index = str(crime.index) + 's'
crime_decade.index.name = "Decade Start"
crime_decade.head()
```

### Step 9. What is the most dangerous decade to live in the US?

```python
# The 1990s
# Alternatively, instead of measuring total crime, we could have measure total crime per mean population. 
# The code in the cell below computes this measure instead
crime_decade.sort_values("Total", ascending=False)
```

```python
column_agg = {col: ('mean' if col == 'Population' else 'sum') for col in crime2.columns}
column_agg
```

```python
crime2_decade = crime2.groupby(decade).aggregate(column_agg)
crime2_decade
```

```python
# going off of crime rate, the most dangerous decade was actually the 1980s
crime2_decade['Crime rate'] = crime2_decade['Total']/ crime2_decade['Population'] 
crime2_decade.sort_values('Crime rate', ascending=False)
```

```python
# using resample; but notice this gives the start date
crime2_decade = crime2.resample('10AS').sum()
crime2_decade['Population'] = crime2['Population'].resample('10AS').max()
crime2_decade
```

```python
crime2_decade.index = crime2_decade.index.year
```
